# 畫出最常用(前15)的語言

```python
data = pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv")
data.head()
```

```python
from collections import Counter
language_counter = Counter()

allLanguage = data["LanguagesWorkedWith"]

for response in allLanguage:
    language_counter.update(response.split(";"))
        
most15language = language_counter.most_common(15)
```

```python
languages = []
popularity = []
for item in most15language:
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages.reverse()
popularity.reverse()
plt.style.use('fivethirtyeight')
plt.barh(languages,popularity)
plt.title("most pkpolar languages")
plt.ylabel("programming languages")
plt.xlabel("number of people who use")
```

```python
plt.style.use('fivethirtyeight')

slices = popularity[0:5]
labels = languages[0:5]
colors = ["#008fd5","#fc4f30","#e5ae37","#6d904f"]
explode = [0,0,0,0.1,0]

plt.pie(slices,
        labels=labels,
        explode=explode,
        startangle=90,
        autopct="%1.1f%%",
        wedgeprops={"edgecolor":"black"})


plt.title("Pie chart")
plt.tight_layout()
plt.show()
```

# Filling Area on Line Plots
    plt.fill_between()

```python
data5 = pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv")
data5.head()
```

```python
ages = data5.Age
dev_salaries = data5.All_Devs
py_salaries = data5.Python
js_salaries = data5.JavaScript

plt.plot(ages,dev_salaries,color = "#444444",
         linestyle="--",label = "All Devs")
plt.plot(ages,py_salaries,label = "Python")

plt.fill_between(ages,py_salaries,dev_salaries,
                 where=(py_salaries>=dev_salaries),
                 alpha = 0.25,label = "之上")

plt.fill_between(ages,py_salaries,dev_salaries,
                 where=(py_salaries<=dev_salaries),
                 color = "red",alpha = 0.25,
                 label = "之下")


plt.legend()
plt.title("median salary (usd) by age")
plt.xlabel("ages")
plt.ylabel("median salary (usd)")
plt.show()
```

```python
from datetime import datetime,timedelta
from matplotlib import dates as mpl_dates
```

    plt.gcf()  #get current figure
    plt.gca()  #get current axis

```python
plt.style.use("seaborn")
dates=[
    datetime(2019,5,24),
    datetime(2019,5,25),
    datetime(2019,5,26),
    datetime(2019,5,27),
    datetime(2019,5,28),
    datetime(2019,5,29),
    datetime(2019,5,30)
]

y = [0,1,2,3,5,5,7]

plt.plot_date(dates,y,linestyle = "solid")
plt.gcf().autofmt_xdate()  #旋轉X座標文字
date_format = mpl_dates.DateFormatter("%b,%d %y")
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```

# plotting live data in real-time
    FuncAnimation(plt.gcf(),animate,interval = 1000)

```python
import random
from itertools import count
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib notebook
plt.style.use("fivethirtyeight")
X_values = []
Y_values = []

index = count()

def animate(i):
    X_values.append(next(index))
    Y_values.append(random.randint(0,5))
    plt.cla()  # 清除axes，即当前figure 中的活动的axes，但其他axes保持不变。
    plt.plot(X_values,Y_values)
    
    
ani = FuncAnimation(plt.gcf(),animate,interval = 1000)
ani
plt.tight_layout()
plt.show()
```

```python
import csv
import random
import time

x_value = 0
total_1 = 1000
total_2 = 1000

fieldnames = ["x_value", "total_1", "total_2"]


with open('data.csv', 'w') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

while True:

    with open('data.csv', 'a') as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        info = {
            "x_value": x_value,
            "total_1": total_1,
            "total_2": total_2
        }

        csv_writer.writerow(info)
        print(x_value, total_1, total_2)

        x_value += 1
        total_1 = total_1 + random.randint(-6, 8)
        total_2 = total_2 + random.randint(-5, 6)

    time.sleep(1)
```

```python
%matplotlib notebook
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()


def animate(i):
    data = pd.read_csv('data.csv')
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']

    plt.cla()

    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')

    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()
```

# Subplots
    plt.subplots(nrows=,ncols=,sharex=)
    fig.savefig("檔名")

```python
data10 = pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/10-Subplots/data.csv")
data10.head()
```

```python
plt.style.use("seaborn")
ages = data10.Age
dev_salaries = data10.All_Devs
py_salaries = data10.Python
js_salaries = data10.JavaScript

fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True)

ax2.plot(ages,py_salaries,label = "Python")
ax2.plot(ages,js_salaries,label = "JS")
ax1.plot(ages,dev_salaries,label = "All",
         color = "#444444",linestyle = "--")

ax1.set_title("median salary")
ax1.set_ylabel("salary")
ax1.legend()


ax2.set_xlabel("age")
ax2.set_ylabel("salary")
ax2.legend()

plt.tight_layout()
plt.show()

fig.savefig("median salary")
```
