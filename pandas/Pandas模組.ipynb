# Tidy Data
    在進行數據分析的時候，資料整理的格式十分重要，基本原則有三
    每一個變數(variable)形成一欄
    每一觀察值(observation)形成一列
    每一種觀測單位形成一個表格
    1. Each variable forms a column.
    2. Each observation forms a row.
    3. Each type of observational unit forms a table.
    
    
    混亂的資料集( messy data)常見的問題有以下幾點
    
    欄標題是數值，而不是變數名稱
    多個變數在同一欄
    變數同時在欄和列
    多種觀察單位存在同一張表格
    單一觀察單位存在多個表格
    1. Column headers are values, not variable names.
    2. Multiple variables are stored in one column.
    3. Variables are stored in both rows and columns.
    4. Multiple types of observational units are stored in the same table.
    5. A single observational unit is stored in multiple tables.

# 1. Column headers are values, not variable names. 

```python
pew = pd.read_csv("https://raw.githubusercontent.com/chendaniely/scipy-2019-pandas/master/data/pew.csv") 
pew
```

# 修正方法 
    .melt(id_var = "不需改變的行名",
          var_name="其他行的名",
          value_name="值的名")

```python
pew_tidy = pew.melt(id_vars = "religion",var_name="income",value_name="count")
pew_tidy
```

```python
billboard = pd.read_csv("https://raw.githubusercontent.com/chendaniely/scipy-2019-pandas/master/data/billboard.csv")
billboard.head()
```

```python
billboard.melt(id_vars=["year","artist","track","time","date.entered"],
               var_name="week",
               value_name="rank")
```

每個artist的平均

```python
billboard   \
    .melt(id_vars=["year","artist","track","time","date.entered"],
          var_name="week",
          value_name="rank")\
    .groupby("artist")["rank"]\
    .mean()
```

# function vectorize

```python
@np.vectorize
def avg_2_mod(x,y):
    if (x == 2):
        return np.nan
    else:
        return (x+y)/2
```

```python
avg_2_mod(df["x"],df["y"])
```
